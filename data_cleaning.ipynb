```python
import json
import pandas as pd

output = []
skipped = []

with open('data2.json', 'r', encoding='utf-8', errors='replace') as f:
    content = f.read()

# Remove outer brackets
content = content.strip()
if content.startswith('['):
    content = content[1:]
if content.endswith(']'):
    content = content[:-1]

# Split line by line
lines = content.split('\n')
print(f"Total lines found: {len(lines)}")

for i, line in enumerate(lines):
    line = line.strip().rstrip(',')
    if not line:
        continue
    try:
        record = json.loads(line)
        output.append(record)
    except json.JSONDecodeError:
        skipped.append(i)
        continue  # skip bad line and continue вњ…

    if i % 10000 == 0:
        print(f"Processed {i} lines... Recovered: {len(output)}")

print(f"\nвњ… Recovered: {len(output)}")
print(f"вќЊ Skipped lines: {len(skipped)}")
print(f"Skipped at lines: {skipped[:10]}")
```

```python
df = pd.DataFrame(output)
```

```python
import csv
df.to_csv("raw_dataset.csv",index=False,encoding="utf-8",quoting=csv.QUOTE_ALL)
```
